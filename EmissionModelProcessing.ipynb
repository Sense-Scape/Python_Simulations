# Emission Modelling

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 5, 1000)

# PSD
Y = 67.62578 + 154.6438*np.power(x,1) - -78.35566* +np.power(x,2) +  11.08792*np.power(x,3)
Y = np.power(10,Y)
Y = np.concatenate((np.flip(Y), Y))

print(Y)
y = np.fft.ifft(Y)

plt.plot(Y)
plt.plot(y)
plt.show()
```
